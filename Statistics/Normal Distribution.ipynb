## Discrete Random Variables
> *A discrete random variable is one which may take on only a countable number of distinct values and thus can be quantified. For example, a variable over a non-empty finite set of integers is discrete, but one over the set of real numbers is continuous. Colloquially, an event which can happen only finitely many times (such as the obtaining of a specific value after repeated rolls of a die) is said to be discrete; whereas an event which can happen an uncountably infinite number of times (such as the height of a person) is said to be continuous.*

[!Wikipedia](https://en.wikipedia.org/wiki/Discrete_random_variable)

If $X$ is a discrete random variable then its range $R_X$ is a countable set, so, we can list the elements in $R_X$ as $\{x_1,x_2,x_3,...\}$. In other words, we can write $$R_X=\{x_1,x_2,x_3,...\}.$$. 
Note that here $x_1,x_2,x_3,...$ are possible values of the random variable $X$. While random variables are usually denoted by capital letters, to represent the numbers in the range we usually use lowercase letters such as $x,x_1,y,z$, etc. For a discrete random variable $X$, we are interested in knowing the probabilities of $X=x_k$
     

### Probability Mass Function
> Let $X$ be a discrete random variable with range $R_X=\{x_1,x_2,x_3,...\}$ (finite or countably infinite). The function
$$P_X(x_k)=P(X=x_k)\text{, for k = 1,2,3,...,}$$
is called the *probability mass function (PMF) of $X$*


> *In probability and statistics, a probability mass function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value. Sometimes it is also known as the discrete density function. The probability mass function is often the primary means of defining a discrete probability distribution, and such functions exist for either scalar or multivariate random variables whose domain is discrete.*

[!Wikipedia](https://en.wikipedia.org/wiki/Probability_mass_function)  



#### Example 1
I toss a fair coin twice, and let $X$ be defined as the number of heads I observe.
1. What is the range of $X$?
Here, our sample space is given by
$$S={HH,HT,TH,TT}.$$
The number of heads will be 0, 1 or 2. Thus, the range of $X$ is
$$R_X={0,1,2}.$$

2. What is the probability mass function of $X$?
We have $$P_X(0)=P(X=0)=P(TT)=\frac{1}{4}$$$$P_X(1)=P(X=1)=P(HT)+P(TH)=\frac{1}{2}$$$$P_X(2)=P(X=2)=P(HH)=\frac{1}{4}$$ Thus, the probability mass function of $X$ is $$P_X(X)=\begin{cases}\frac{1}{4},&\text{if }x=0\\\frac{1}{2},&\text{if }x=1\\\frac{1}{4},&\text{if }x=2\\0,&\text{otherwise}\end{cases}$$ 

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np


xk = list(range(5))
pk = (0.0, 0.25, 0.5, 0.25, 0.0)
custm = stats.rv_discrete(name='custm', values=(xk, pk))

fig, ax = plt.subplots(1, 1)
ax.plot(xk, custm.pmf(xk), 'ro', ms=8, mec='r')
ax.vlines(xk, 0, custm.pmf(xk), colors='r', linestyles='-', lw=2)
ax.set_xticks(xk)
plt.title('PMF for random Variable X in Example 1')
plt.ylabel('$P_X(X)$')
plt.show()
```

### Independent Random Variables
> *Two random variables X and Y are independent if their joint probability mass function (pmf) or joint probability density function (pdf) can be factored into the product of two functions, one involving only X and the other involving only Y.*

[!Wikipedia](https://en.wikipedia.org/wiki/Independence_(probability_theory))   

In other words, $$P(X=x,Y=y)=P(X=x)P(Y=y)$$ for all $x$ and $y$ in the range of $X$ and $Y$ respectively.



### Special DIstributions
There are some specific distributions that are used over and over in practice, thus they have been given special names. There is a random experiment behind each of these distributions. Since these random experiments model a lot of real life phenomenon, these special distributions are used frequently in different applications. 

#### Bernoulli Distribution
A Bernoulli random variable is a random variable that can only take two possible values, usually 0
 and 1. This random variable models random experiments that have two possible outcomes, sometimes referred to as "success" and "failure." Here are some examples:

* You take a pass-fail exam. You either pass (resulting in X=1) or fail (resulting in X=0).
* You toss a coin. The outcome is either heads or tails.
* A child is born. The gender is either male or female.

> A random variable X is said to be a Bernoulli random variable with parameter p, shown as X∼Bernoulli(p)
, if its PMF is given by $$P_X(x)=\begin{cases}p,&\text{if }x=1\\1-p,&\text{if }x=0\\0,&\text{otherwise}\end{cases}$$ where $0<p<1$.


```python
# Import the required libraries
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Instance of Bernoulli distribution with parameter p = 0.8
bd=bernoulli(0.8)
# Outcome of random variable either 0 or 1
xk=[-1,0,1,2] 

fig, ax = plt.subplots(1, 1)
ax.plot(xk, bd.pmf(xk), 'ro', ms=8, mec='r')
ax.vlines(xk, 0, bd.pmf(xk), colors='r', linestyles='-', lw=2)
ax.set_xticks(xk)
plt.title('PMF for Bernoulli random Variable $X ∼ Bernoulli(p)$')
plt.ylabel('$P_X(X)$')
plt.xlabel('Values of random variable x (0, 1)')
plt.show()
```

#### Geometric Distribution
The geometric distribution is a discrete probability distribution that models the number of trials that must be run in order to achieve success. It is often used to model the number of trials up to and including the first success. E.g., suppose that I have a coin with $P(H)=p$ and $P(T)=1-p$
. I toss the coin until I observe the first heads. We define X as the total number of coin tosses in this experiment. Then X is said to have geometric distribution with parameter p. In other words, you can think of this experiment as repeating independent Bernoulli trials until observing the first success.

> A random variable X is said to be a geometric random variable with parameter p, shown as X∼Geometric(p), if its PMF is given by $$P_X(x)=\begin{cases}p(1-p)^{x-1},&\text{if }x=1,2,3,...\\0,&\text{otherwise}\end{cases}$$ where $0<p<1$.

```python
# Import the required libraries
from scipy.stats import geom
import matplotlib.pyplot as plt

# Instance of Geometric distribution with parameter p = 0.3
gd=geom(0.3)

xk=list(range(1,20))

fig, ax = plt.subplots(1, 1)
ax.plot(xk, gd.pmf(xk), 'ro', ms=8, mec='r')
ax.vlines(xk, 0, gd.pmf(xk), colors='r', linestyles='-', lw=2)
ax.set_xticks(xk)
plt.title('PMF for Geometric random Variable $X ∼Geometric(0.3)$')
plt.ylabel('$P_X(X)$')
plt.xlabel('Values of random variable x (1, 2, 3, ...)')
plt.show()
```

#### Binomial Distribution
The binomial distribution is a discrete probability distribution that models the number of successes in a sequence of n independent experiments. It is often used to model the number of successes in a sample of size n drawn with replacement from a population of size N. E.g., suppose that I have a coin with $P(H)=p$ and $P(T)=1-p$ and I toss the coin n times. We define X as the total number of heads in this experiment. Then X is said to have binomial distribution with parameters n and p. In other words, you can think of this experiment as repeating independent Bernoulli trials n times.

> A random variable X is said to be a binomial random variable with parameters n and p, shown as X∼Binomial(n,p), if its PMF is given by $$P_X(x)=\begin{cases}\binom{n}{x}p^x(1-p)^{n-x},&\text{if }x=0,1,2,...,n\\0,&\text{otherwise}\end{cases}$$ where $0<p<1$ and $n$ is a positive integer.



```python
# Import the required libraries
from scipy.stats import binom
import matplotlib.pyplot as plt

# Instance of Binomial distribution with parameter n = 10 and p = 0.3
bd=binom(10,0.3)
xk=list(range(11))

bd1 = binom(20, 0.6)
xk2 = np.arange(0, 21)

fig, ax = plt.subplots(2, 1)

ax[0].plot(xk, bd.pmf(xk), 'ro', ms=8, mec='r')
ax[0].vlines(xk, 0, bd.pmf(xk), colors='r', linestyles='-', lw=2)
ax[0].set_xticks(xk)
ax[0].set_title('PMF for Binomial random Variable $X ∼Binomial(10,0.3)$')
ax[0].set_ylabel('$P_X(X)$')
ax[0].set_xlabel('Values of random variable x (0, 1, 2, ..., 10)')
ax[1].plot(xk2, bd1.pmf(xk2), 'ro', ms=8, mec='r')
ax[1].vlines(xk2, 0, bd1.pmf(xk2), colors='r', linestyles='-', lw=2)
ax[1].set_xticks(xk2)
ax[1].set_title('PMF for Binomial random Variable $X ∼Binomial(20,0.6)$')
ax[1].set_ylabel('$P_X(X)$')
ax[1].set_xlabel('Values of random variable x (0, 1, 2, ..., 20)')
plt.subplots_adjust(hspace=0.5)
plt.show()
```

##### Binomial random variable as a sum of Bernoulli random variables
Here is a useful way of thinking about a binomial random variable. Note that a Binomial(n,p)
 random variable can be obtained by n independent coin tosses. If we think of each coin toss as a Bernoulli(p) random variable, the Binomial(n,p) random variable is a sum of n independent Bernoulli(p) random variables. This is stated more precisely in the following lemma.
 
**Lemma** 
If $X_1,X_2,...,X_n$ are independent Bernoulli(p) random variables, then the random variable $X$ defined by $X=X_1+X_2+...+X_n$ has a Binomial(n,p) distribution.

To generate a random variable $X∼Binomial(n,p)$, we can toss a coin n times and count the number of heads. Counting the number of heads is exactly the same as finding $X_1+X_2+...+X_n$, where each $X_i$
 is equal to one if the corresponding coin toss results in heads and zero otherwise. This interpretation of binomial random variables is sometimes very helpful. 

#### Poisson Distribution

The Poisson distribution is one of the most widely used probability distributions. It is usually used in scenarios where we are counting the occurrences of certain events in an interval of time or space. In practice, it is often an approximation of a real-life random variable. Here is an example of a scenario where a Poisson random variable might be used. Suppose that we are counting the number of customers who visit a certain store from 1pm to 2pm. Based on data from previous days, we know that on average $\lambda=15$ customers visit the store. Of course, there will be more customers some days and fewer on others. Here, we may model the random variable $X$ showing the number customers as a Poisson random variable with parameter $\lambda=15$.

> A random variable X is said to be a Poisson random variable with parameter $\lambda$, shown as $X∼Poisson(\lambda)$, if its PMF is given by $$P_X(x)=\begin{cases}\frac{\lambda^xe^{-\lambda}}{x!},&\text{if }x=0,1,2,...\\0,&\text{otherwise}\end{cases}$$ where $\lambda>0$.

### Cumulative Distribution Function

> *In probability theory and statistics, the cumulative distribution function (CDF) of a real-valued random variable X, or just distribution function of X, evaluated at x, is the probability that X will take a value less than or equal to x. In the case of a continuous distribution, it gives the area under the probability density function from minus infinity to x. Cumulative distribution functions are also used to specify the distribution of multivariate random variables.*

[!Wikipedia](https://en.wikipedia.org/wiki/Cumulative_distribution_function)


#### Example 2

## Multivariate Normal Distribution
To find the values of vectors for a given probability in a multivariate distribution (e.g., bivariate normal), we need to solve for the vector that corresponds to a given cumulative probability. This can be approached using numerical methods like optimization because the inverse CDF for multivariate distributions is not directly available in most statistical libraries.

Let's consider a bivariate normal distribution as an example. We will use the `scipy.optimize` module to find the vector values for a given cumulative probability 
P.

```python
import numpy as np
import scipy.stats as stats
import scipy.optimize as optimize

# Define the parameters for the bivariate normal distribution
mean = [0, 0]  # Example means
cov = [[1, 0.5], [0.5, 1]]  # Example covariance matrix, implying some correlation

# The target probability
P = 0.5

# Define the objective function to minimize: the difference between the CDF and the target probability
def objective(x, mean, cov, target_P):
    return (stats.multivariate_normal(mean, cov).cdf(x) - target_P)**2

# Initial guess for the optimization
x0 = np.zeros(len(mean))

# Use scipy.optimize to find the vector x for which the CDF is close to P
result = optimize.minimize(objective, x0, args=(mean, cov, P))

# The vector values that correspond to the cumulative probability P
vector_x = result.x

print(f"The vector values for which the CDF is approximately {P} are {vector_x}")
```

## Bivariate Normal PDF

The contours join points on the surface that have the same height. For any arbitrary normal, the contours are __elliptical with the center at the mean__. 
The width and orientation of the ellipses are determined by the __variance-covariance matrix__.
The color represents the __height of the density__ at any given location. 
Changing the __mean__ will move the __center of ellipses__ to a new location, without altering the shape.
Changing the __variance-covariance matrix__ changes the __shape__ of the ellipses. In the special case where the off-diagonal entry, or covariance, equals zero and the variances are equal, the ellipses become circles.

The contours are perpendicular to the eigenvectors of the covariance matrix. 
The eigenvectors are the directions of the principal axes of the ellipse, and the eigenvalues are the variances of the data in these directions. The larger the eigenvalue, the longer the ellipse in the direction of the corresponding eigenvector.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters for the bivariate normal distribution
mu_X, mu_Y = 0, 0  # Means
sigma_X, sigma_Y = 1, 1  # Standard deviations
rho = 0.5  # Correlation coefficient
cov = [[sigma_X**2, rho * sigma_X * sigma_Y], [rho * sigma_X * sigma_Y, sigma_Y**2]]  # Covariance matrix

# Define the bivariate normal distribution
bivariate_normal_dist = multivariate_normal([mu_X, mu_Y], cov)

# Create a grid of points
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Compute the PDF on the grid
pdf = bivariate_normal_dist.pdf(pos)

# Compute the CDF on the grid
cdf = bivariate_normal_dist.cdf(pos)

# Generate samples from the distribution
samples = bivariate_normal_dist.rvs(1000)
print (samples.shape)

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.contour(x, y, pdf, cmap='viridis')
plt.colorbar()

plt.contour(x, y, cdf, cmap='autumn')

plt.scatter(samples[:,0], samples[:,1], color='red', s=0.75)
# plt.hold()

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bivariate Normal PDF')
plt.show()
```

```python
#PDF, CDF and Samples of normal distribution
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set parameters for the normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Generate samples from the normal distribution
samples = np.random.normal(mu, sigma, 1000)

# Create an array of x values for plotting the PDF and CDF
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Calculate the 95th percentile of the normal distribution
percentile_95 = norm.ppf(0.95, mu, sigma)

# Plotting with the 95th percentile line
plt.figure(figsize=(12, 8))

# Plot PDF
plt.plot(x, pdf, label='PDF', color='blue')

# Plot CDF
plt.plot(x, cdf, label='CDF', color='red')

# Plot the samples on the x-axis
plt.scatter(samples, np.zeros_like(samples), marker='|', color='green', alpha=0.7, label='Samples')

# Add a line for the 95th percentile
plt.axvline(percentile_95, color='purple', linestyle='--', label=f'95th Percentile ({percentile_95:.2f})')

plt.title('Normal Distribution: PDF, CDF, Samples, and 95th Percentile')
plt.xlabel('Value')
plt.ylabel('Probability / Cumulative Probability')
plt.legend()
plt.grid(True)
plt.show()
```

## Bivariate PDF, Samples and 95 percentile
The graph illustrates the PDF of a bivariate normal distribution along with samples drawn from it. Two lines represent the 95th percentile for each dimension: the red dashed line for the X dimension and the blue dashed line for the Y dimension. These lines indicate the value below which 95% of the samples fall in their respective dimensions.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the mean and covariance of the bivariate normal distribution
mean = [0, 0]
covariance = [[1, 0.5], [0.5, 1]]  # Covariance matrix with some correlation

# Create a bivariate normal distribution object
bivariate_normal = multivariate_normal(mean, covariance)

# Generate grid points for evaluating the PDF
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Calculate the PDF
pdf = bivariate_normal.pdf(pos)

# Generate samples from the bivariate normal distribution
samples = bivariate_normal.rvs(size=1000)

# Calculate the 95th percentile threshold for each variable independently
percentile_95_x = np.percentile(samples[:, 0], 95)
percentile_95_y = np.percentile(samples[:, 1], 95)

# Plotting
plt.figure(figsize=(12, 8))

# Plot the PDF
plt.contour(x, y, pdf, cmap='viridis')

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], color='white', s=10, alpha=0.3)

# Add lines for the 95th percentile in each dimension
plt.axvline(percentile_95_x, color='red', linestyle='--', label=f'95th Percentile X ({percentile_95_x:.2f})')
plt.axhline(percentile_95_y, color='blue', linestyle='--', label=f'95th Percentile Y ({percentile_95_y:.2f})')

plt.title('Bivariate Normal Distribution with Samples and 95th Percentile')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='PDF Value')
plt.legend()
plt.grid(True)
plt.show()
```

## Adding CDF to the Bivariate PDF
Adding a CDF (Cumulative Distribution Function) plot for a bivariate normal distribution directly on a 2D plot like this one can be challenging, as the CDF in a multivariate context is a multidimensional function. However, we can illustrate the concept of cumulative probability by showing contour lines of the CDF, which represent levels of cumulative probability. The contour lines are perpendicular to the principal axes of the distribution and indicate the probability of observing a value below a certain threshold at each point in the plot. 
To properly add CDF information to the plot, one would need to compute the cumulative probability over the grid, which can be complex and computationally intensive. Thus, the above approach is a simplified attempt to illustrate where certain levels of probability density might correspond to cumulative levels, but it does not accurately represent the CDF.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the mean and covariance of the bivariate normal distribution
mean = [0, 0]
covariance = [[1, 0.5], [0.5, 1]]  # Covariance matrix with some correlation

# Create a bivariate normal distribution object
bivariate_normal = multivariate_normal(mean, covariance)

# Generate grid points for evaluating the PDF and CDF
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Calculate the PDF and CDF
pdf = bivariate_normal.pdf(pos)
cdf = bivariate_normal.cdf(pos)

# Generate samples from the bivariate normal distribution
samples = bivariate_normal.rvs(size=1000)

# Calculate the 95th percentile threshold for each variable independently
percentile_95_x = np.percentile(samples[:, 0], 95)
percentile_95_y = np.percentile(samples[:, 1], 95)

# Define PDF levels for contour lines and annotations
cdf_levels = np.linspace(np.min(cdf), np.max(cdf), num=20)

# Plotting
plt.figure(figsize=(12, 8))

# Plot the PDF
plt.contour(x, y, pdf, cmap='viridis')

# Plot CDF contours
contour_lines = plt.contour(x, y, cdf, levels = cdf_levels, cmap='viridis', linestyles='dashed', linewidths=2) 
contour_lines.clabel(cdf_levels, inline=True, fontsize=12, fmt='%.2f')

# Plot the samples
plt.scatter(samples[:, 0], samples[:, 1], color='red', s=2, alpha=0.3)

# Add lines for the 95th percentile in each dimension
plt.axvline(percentile_95_x, color='red', linestyle='--', label=f'95th Percentile X ({percentile_95_x:.2f})')
plt.axhline(percentile_95_y, color='blue', linestyle='--', label=f'95th Percentile Y ({percentile_95_y:.2f})')

plt.title('Bivariate Normal Distribution: PDF, CDF Contours, Samples, and 95th Percentile')
plt.xlabel('X')
plt.ylabel('Y')
plt.colorbar(label='PDF Value')
plt.legend()
plt.grid(True)
plt.show()
```

### Sum of two normal variables
The sum of two independent normal random variables is also a normal random variable. If $X∼N(μ_1,σ_1^2)$ and $Y∼N(μ_2,σ_2^2)$ are independent normal random variables, then $X+Y∼N(μ_1+μ_2,σ_1^2+σ_2^2)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distributions
mu1, sigma1 = 0, 1
mu2, sigma2 = 1, 2

# Create the normal distributions
normal1 = norm(mu1, sigma1)
normal2 = norm(mu2, sigma2)

# Generate a grid of x values
x = np.linspace(-10, 10, 1000)

# Calculate the PDFs of the normal distributions
pdf1 = normal1.pdf(x)
pdf2 = normal2.pdf(x)

# Calculate the PDF of the sum of the two normal distributions
pdf_sum = np.convolve(pdf1, pdf2, 'same')

# Plot the PDFs
plt.figure(figsize=(12, 8))
plt.plot(x, pdf1, label='N(0, 1)', color='blue')
plt.plot(x, pdf2, label='N(1, 2)', color='red')
plt.plot(x, pdf_sum, label='Sum PDF', color='green', linestyle='--')

plt.title('PDF of Sum of Two Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Density')

plt.legend()
plt.grid(True)
plt.show()

```

### Sum of two independent normal random variables
The sum of two independent normal random variables is also a normal random variable. If $X∼N(μ_1,σ_1^2)$ and $Y∼N(μ_2,σ_2^2)$ are independent normal random variables, then $X+Y∼N(μ_1+μ_2,σ_1^2+σ_2^2)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd

# Parameters for the normal distributions
mu1, sigma1 = 10, 1
mu2, sigma2 = 100, 20

# Create the normal distributions
vm1_dist = norm(mu1, sigma1)
vm2_dist = norm(mu2, sigma2)
agg_dist = norm(mu1 + mu2, np.sqrt(sigma1**2 + sigma2**2))


df = pd.DataFrame({
    'Name': ['VM1', 'VM2', 'VM1+VM2', 'Aggregated'],
    'Organic': [vm1_dist.median(), vm2_dist.median(), vm1_dist.median()+vm2_dist.median(), agg_dist.median()],
    'Safety Stock': [1.64*vm1_dist.std(), 1.64*vm2_dist.std(), 1.64*vm1_dist.std()+ 1.64*vm2_dist.std(),1.64*agg_dist.std()],
})

df['Total'] = df['Organic'] + df['Safety Stock']
df['Total'] = df['Total'].round(2)
display(df)

# Generate samples from the normal distributions
vm1 = vm1_dist.rvs(size=1000)
vm2 = vm2_dist.rvs(size=1000)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(0, 10, 400)
y = np.linspace(0, 10, 400)

# Create a meshgrid
X, Y = np.meshgrid(x, y)

# Calculate Z based on the equation x + y - sqrt(x^2 + y^2)
Z_expression = ((X + Y - np.sqrt(X**2 + Y**2))/(X+Y))*100

# Create a 3D plot for the expression
fig_exp = plt.figure()
ax_exp = fig_exp.add_subplot(111, projection='3d')
surface_exp = ax_exp.plot_surface(X, Y, Z_expression, cmap='coolwarm')
fig_exp.colorbar(surface_exp, shrink=0.5, aspect=5)
plt.title('$x + y - \sqrt{x^2 + y^2}$ 3D Plot')
ax_exp.set_xlabel('X axis')
ax_exp.set_ylabel('Y axis')
ax_exp.set_zlabel('$x + y - \sqrt{x^2 + y^2}$')

# Create a contour plot for the same expression
plt.figure()
contour_exp = plt.contour(X, Y, Z_expression, levels=50, cmap='coolwarm')
plt.colorbar(contour_exp)
plt.title('Contour Plot of $x + y - \sqrt{x^2 + y^2}$')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```
