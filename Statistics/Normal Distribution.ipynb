## Discrete Random Variables
> *A discrete random variable is one which may take on only a countable number of distinct values and thus can be quantified. For example, a variable over a non-empty finite set of integers is discrete, but one over the set of real numbers is continuous. Colloquially, an event which can happen only finitely many times (such as the obtaining of a specific value after repeated rolls of a die) is said to be discrete; whereas an event which can happen an uncountably infinite number of times (such as the height of a person) is said to be continuous.*

[!Wikipedia](https://en.wikipedia.org/wiki/Discrete_random_variable)

If $X$ is a discrete random variable then its range $R_X$ is a countable set, so, we can list the elements in $R_X$ as $\{x_1,x_2,x_3,...\}$. In other words, we can write $$R_X=\{x_1,x_2,x_3,...\}.$$. 
Note that here $x_1,x_2,x_3,...$ are possible values of the random variable $X$. While random variables are usually denoted by capital letters, to represent the numbers in the range we usually use lowercase letters such as $x,x_1,y,z$, etc. For a discrete random variable $X$, we are interested in knowing the probabilities of $X=x_k$
     

### Probability Mass Function
> Let $X$ be a discrete random variable with range $R_X=\{x_1,x_2,x_3,...\}$ (finite or countably infinite). The function
$$P_X(x_k)=P(X=x_k)\text{, for k = 1,2,3,...,}$$
is called the *probability mass function (PMF) of $X$*


> *In probability and statistics, a probability mass function (PMF) is a function that gives the probability that a discrete random variable is exactly equal to some value. Sometimes it is also known as the discrete density function. The probability mass function is often the primary means of defining a discrete probability distribution, and such functions exist for either scalar or multivariate random variables whose domain is discrete.*

[!Wikipedia](https://en.wikipedia.org/wiki/Probability_mass_function)  



#### Example 1
I toss a fair coin twice, and let $X$ be defined as the number of heads I observe.
1. What is the range of $X$?
Here, our sample space is given by
$$S={HH,HT,TH,TT}.$$
The number of heads will be 0, 1 or 2. Thus, the range of $X$ is
$$R_X={0,1,2}.$$

2. What is the probability mass function of $X$?
We have $$P_X(0)=P(X=0)=P(TT)=\frac{1}{4}$$$$P_X(1)=P(X=1)=P(HT)+P(TH)=\frac{1}{2}$$$$P_X(2)=P(X=2)=P(HH)=\frac{1}{4}$$ Thus, the probability mass function of $X$ is $$P_X(X)=\begin{cases}\frac{1}{4},&\text{if }x=0\\\frac{1}{2},&\text{if }x=1\\\frac{1}{4},&\text{if }x=2\\0,&\text{otherwise}\end{cases}$$ 

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np


xk = list(range(5))
pk = (0.0, 0.25, 0.5, 0.25, 0.0)
custm = stats.rv_discrete(name='custm', values=(xk, pk))

fig, ax = plt.subplots(1, 1)
ax.plot(xk, custm.pmf(xk), 'ro', ms=8, mec='r')
ax.vlines(xk, 0, custm.pmf(xk), colors='r', linestyles='-', lw=2)
ax.set_xticks(xk)
plt.title('PMF for random Variable X in Example 1')
plt.ylabel('$P_X(X)$')
plt.show()
```

### Independent Random Variables
> *Two random variables X and Y are independent if their joint probability mass function (pmf) or joint probability density function (pdf) can be factored into the product of two functions, one involving only X and the other involving only Y.*

[!Wikipedia](https://en.wikipedia.org/wiki/Independence_(probability_theory))   

In other words, $$P(X=x,Y=y)=P(X=x)P(Y=y)$$ for all $x$ and $y$ in the range of $X$ and $Y$ respectively.



### Special DIstributions
There are some specific distributions that are used over and over in practice, thus they have been given special names. There is a random experiment behind each of these distributions. Since these random experiments model a lot of real life phenomenon, these special distributions are used frequently in different applications. 

#### Bernoulli Distribution
A Bernoulli random variable is a random variable that can only take two possible values, usually 0
 and 1. This random variable models random experiments that have two possible outcomes, sometimes referred to as "success" and "failure." Here are some examples:

* You take a pass-fail exam. You either pass (resulting in X=1) or fail (resulting in X=0).
* You toss a coin. The outcome is either heads or tails.
* A child is born. The gender is either male or female.

> A random variable X is said to be a Bernoulli random variable with parameter p, shown as Xâˆ¼Bernoulli(p)
, if its PMF is given by $$P_X(x)=\begin{cases}p,&\text{if }x=1\\1-p,&\text{if }x=0\\0,&\text{otherwise}\end{cases}$$ where $0<p<1$.


```python
# Import the required libraries
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Instance of Bernoulli distribution with parameter p = 0.8
bd=bernoulli(0.8)
# Outcome of random variable either 0 or 1
xk=[-1,0,1,2] 

fig, ax = plt.subplots(1, 1)
ax.plot(xk, bd.pmf(xk), 'ro', ms=8, mec='r')
ax.vlines(xk, 0, bd.pmf(xk), colors='r', linestyles='-', lw=2)
ax.set_xticks(xk)
plt.title('PMF for Bernoulli random Variable $X âˆ¼ Bernoulli(p)$')
plt.ylabel('$P_X(X)$')
plt.xlabel('Values of random variable x (0, 1)')
plt.show()
```

#### Geometric Distribution
The geometric distribution is a discrete probability distribution that models the number of trials that must be run in order to achieve success. It is often used to model the number of trials up to and including the first success. E.g., suppose that I have a coin with $P(H)=p$ and $P(T)=1-p$
. I toss the coin until I observe the first heads. We define X as the total number of coin tosses in this experiment. Then X is said to have geometric distribution with parameter p. In other words, you can think of this experiment as repeating independent Bernoulli trials until observing the first success.

> A random variable X is said to be a geometric random variable with parameter p, shown as Xâˆ¼Geometric(p), if its PMF is given by $$P_X(x)=\begin{cases}p(1-p)^{x-1},&\text{if }x=1,2,3,...\\0,&\text{otherwise}\end{cases}$$ where $0<p<1$.

```python
# Import the required libraries
from scipy.stats import geom
import matplotlib.pyplot as plt

# Instance of Geometric distribution with parameter p = 0.3
gd=geom(0.3)

xk=list(range(1,20))

fig, ax = plt.subplots(1, 1)
ax.plot(xk, gd.pmf(xk), 'ro', ms=8, mec='r')
ax.vlines(xk, 0, gd.pmf(xk), colors='r', linestyles='-', lw=2)
ax.set_xticks(xk)
plt.title('PMF for Geometric random Variable $X âˆ¼Geometric(0.3)$')
plt.ylabel('$P_X(X)$')
plt.xlabel('Values of random variable x (1, 2, 3, ...)')
plt.show()
```

#### Binomial Distribution
The binomial distribution is a discrete probability distribution that models the number of successes in a sequence of n independent experiments. It is often used to model the number of successes in a sample of size n drawn with replacement from a population of size N. E.g., suppose that I have a coin with $P(H)=p$ and $P(T)=1-p$ and I toss the coin n times. We define X as the total number of heads in this experiment. Then X is said to have binomial distribution with parameters n and p. In other words, you can think of this experiment as repeating independent Bernoulli trials n times.

> A random variable X is said to be a binomial random variable with parameters n and p, shown as Xâˆ¼Binomial(n,p), if its PMF is given by $$P_X(x)=\begin{cases}\binom{n}{x}p^x(1-p)^{n-x},&\text{if }x=0,1,2,...,n\\0,&\text{otherwise}\end{cases}$$ where $0<p<1$ and $n$ is a positive integer.



```python
# Import the required libraries
from scipy.stats import binom
import matplotlib.pyplot as plt

# Instance of Binomial distribution with parameter n = 10 and p = 0.3
bd=binom(10,0.3)
xk=list(range(11))

bd1 = binom(20, 0.6)
xk2 = np.arange(0, 21)

fig, ax = plt.subplots(2, 1)

ax[0].plot(xk, bd.pmf(xk), 'ro', ms=8, mec='r')
ax[0].vlines(xk, 0, bd.pmf(xk), colors='r', linestyles='-', lw=2)
ax[0].set_xticks(xk)
ax[0].set_title('PMF for Binomial random Variable $X âˆ¼Binomial(10,0.3)$')
ax[0].set_ylabel('$P_X(X)$')
ax[0].set_xlabel('Values of random variable x (0, 1, 2, ..., 10)')
ax[1].plot(xk2, bd1.pmf(xk2), 'ro', ms=8, mec='r')
ax[1].vlines(xk2, 0, bd1.pmf(xk2), colors='r', linestyles='-', lw=2)
ax[1].set_xticks(xk2)
ax[1].set_title('PMF for Binomial random Variable $X âˆ¼Binomial(20,0.6)$')
ax[1].set_ylabel('$P_X(X)$')
ax[1].set_xlabel('Values of random variable x (0, 1, 2, ..., 20)')
plt.subplots_adjust(hspace=0.5)
plt.show()
```

##### Binomial random variable as a sum of Bernoulli random variables
Here is a useful way of thinking about a binomial random variable. Note that a Binomial(n,p)
 random variable can be obtained by n independent coin tosses. If we think of each coin toss as a Bernoulli(p) random variable, the Binomial(n,p) random variable is a sum of n independent Bernoulli(p) random variables. This is stated more precisely in the following lemma.
 
**Lemma** 
If $X_1,X_2,...,X_n$ are independent Bernoulli(p) random variables, then the random variable $X$ defined by $X=X_1+X_2+...+X_n$ has a Binomial(n,p) distribution.

To generate a random variable $Xâˆ¼Binomial(n,p)$, we can toss a coin n times and count the number of heads. Counting the number of heads is exactly the same as finding $X_1+X_2+...+X_n$, where each $X_i$
 is equal to one if the corresponding coin toss results in heads and zero otherwise. This interpretation of binomial random variables is sometimes very helpful. 

#### Poisson Distribution

The Poisson distribution is one of the most widely used probability distributions. It is usually used in scenarios where we are counting the occurrences of certain events in an interval of time or space. In practice, it is often an approximation of a real-life random variable. Here is an example of a scenario where a Poisson random variable might be used. Suppose that we are counting the number of customers who visit a certain store from 1pm to 2pm. Based on data from previous days, we know that on average $\lambda=15$ customers visit the store. Of course, there will be more customers some days and fewer on others. Here, we may model the random variable $X$ showing the number customers as a Poisson random variable with parameter $\lambda=15$.

> A random variable X is said to be a Poisson random variable with parameter $\lambda$, shown as $Xâˆ¼Poisson(\lambda)$, if its PMF is given by $$P_X(x)=\begin{cases}\frac{\lambda^xe^{-\lambda}}{x!},&\text{if }x=0,1,2,...\\0,&\text{otherwise}\end{cases}$$ where $\lambda>0$.


```python
# Import the required libraries
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Instance of Poisson distribution with parameter lambda = 1, 5, 1
pd1=poisson(1)
pd5=poisson(5)
pd10=poisson(10)

xk=list(range(21))

fig, ax = plt.subplots(3, 1, sharex=True, figsize=(8, 10))
ax[0].plot(xk, pd1.pmf(xk), 'ro', ms=8, mec='r')
ax[0].vlines(xk, 0, pd1.pmf(xk), colors='r', linestyles='-', lw=2)
# ax[0].set_xticks(xk)
ax[0].set_title('PMF for Poisson random Variable $X âˆ¼Poisson(1)$')
ax[0].set_ylabel('$P_X(X)$')
# ax[0].set_xlabel('Values of random variable x (0, 1, 2, ..., 20)')
ax[1].plot(xk, pd5.pmf(xk), 'ro', ms=8, mec='r')
ax[1].vlines(xk, 0, pd5.pmf(xk), colors='r', linestyles='-', lw=2)
# ax[1].set_xticks(xk)
ax[1].set_title('PMF for Poisson random Variable $X âˆ¼Poisson(5)$')
ax[1].set_ylabel('$P_X(X)$')
# ax[1].set_xlabel('Values of random variable x (0, 1, 2, ..., 20)')
ax[2].plot(xk, pd10.pmf(xk), 'ro', ms=8, mec='r')
ax[2].vlines(xk, 0, pd10.pmf(xk), colors='r', linestyles='-', lw=2)
ax[2].set_xticks(xk)
ax[2].set_title('PMF for Poisson random Variable $X âˆ¼Poisson(10)$')
ax[2].set_ylabel('$P_X(X)$')
ax[2].set_xlabel('Values of random variable x (0, 1, 2, ..., 20)')
plt.subplots_adjust(hspace=0.2)
plt.show()

    

        
```

## Probability Density Function
> *The probability density function (PDF) is a statistical expression that defines a probability distribution (the likelihood of an outcome) for a discrete random variable as opposed to a continuous random variable. When the PDF is graphically portrayed, the area under the curve will indicate the interval in which the variable will fall.*

[!Investopedia](https://www.investopedia.com/terms/p/pdf.asp)

## Normal Distribution Cumulative Distribution Function

```python
## Normal Distribution Probability Density Function

fig, ax = plt.subplots()
x= np.arange(-4,4,0.001)
ax.plot(x, norm.pdf(x))
ax.set_ylim(0,0.45) # range
ax.axhline(y=0.24,xmax=0.61,color='r') # horizontal line
ax.axvline(x=1, ymax=0.53, color='r',alpha=0.5) # vertical line
xplot = ax.plot([1], [0.24], marker='o', markersize=15, color="red") # coordinate point
ax.set_yticks([]) # remove y axis label
ax.set_xticks([]) # remove x axis label
ax.set_xlabel('x',fontsize=20) # set x label
ax.set_ylabel('pdf(x)',fontsize=20,rotation=0) # set y label
ax.xaxis.set_label_coords(0.61, -0.02) # x label coordinate
ax.yaxis.set_label_coords(-0.1, 0.5) # y label coordinate
plt.show()

```

Since norm.pdf returns a PDF value, we can use this function to plot the normal distribution function. We graph a PDF of the normal distribution using `scipy`, `numpy` and `matplotlib`. We use the domain
